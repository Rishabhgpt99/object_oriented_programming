```python
# Inheritance basic
# Parent class

class user:
  def __init__(self):
    self.name="Gaurav"
    self.gender="Male"

  def login(self):
      print('login')

# Child Class
class student(user):
  def enroll(self):
    print('Enroll the course')

u = user()
s= student()

print(s.name)
s.enroll()
s.login()
```

```python
class Car:
  def __init__(self,windows,door,enginetype):
    self.windows=windows
    self.door=door
    self.enginetype=enginetype

  def driving(self):
    print("Car is used for driving")

class Audi(Car):
  def __init__(self,windows,door,enginetype,horsepower):
    super().__init__(windows,door,enginetype)
    self.horsepower=horsepower

audiq7=Audi(4,5,'Diesel',200)
print(audiq7.horsepower)
print(audiq7)
```

```python
print(audiq7.door)
audiq7.driving()
```
