```python
# Method Overloading

class Shape:
  def area(self,radius):
    return 3.14*radius*radius

  def area(self,l,b):
    return l*b

s=Shape()
s.area(2)
s.area(3,4)
```

```python
class Shape:
  def area(self,a,b=0):
    if b==0:
      return 3.14*a*a
    else:
      return a*b

s=Shape()
print(s.area(2))
print(s.area(3,4))
```
