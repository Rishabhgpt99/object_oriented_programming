**Encapsulation:**
This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data.

```python
# Access Modifiers
class person:
  def __init__(self,name,age):
    self.name=name  #public
    self.age=age    #public

person=person('ram',45)
```

```python
dir(person)
```

```python
class person:
  def __init__(self,name,age,gender):
    self.__name=name #private
    self.__age=age   #private
    self.gender=gender  #public

```

```python
per=person('ram',45,'male')
dir(per)
```

```python
class person:
  def __init__(self,name,age,gender):
    self._name=name #protected
    self._age=age   #protected
    self.gender=gender

class Employee(person):
  def __init__(self,name,age,gender):
    super().__init__(name,age,gender)

employee=Employee('ram',65,'male')
print(employee._name)
```

**Encapsulation with Getter and Setter Method**

```python
class Person:
  def __init__(self,name,age):
      self.__name=name #private
      self.__age=age

#getter method for name
  def get_name(self):
      return self.__name

#setter method for name
  def set_name(self,name):
      self.__name=name

#getter method for age
  def get_age(self):
      return self.__age

#setter method for age
  def set_age(self,age):
      if age>0:
         self.__age=age
      else:
         print("age cannot be negative")


pers = Person('ram',45)

##Access & modify private variable using getter and setter
print(pers.get_name)
print(pers.get_age)

pers.set_age(40)


pers.set_age(-3)


```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # private
        self.__age = age

    # getter method for name
    def get_name(self):
        return self.__name

    # setter method for name
    def set_name(self, name):
        self.__name = name

    # getter method for age
    def get_age(self):
        return self.__age

    # setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")

pers = Person('ram', 45)

# Access & modify private variable using getter and setter
print(pers.get_name())  # Added parentheses

print(pers.get_age())
print(pers.set_age(-3))
```

# Abstraction
It hides unnecessary code details from the user. Also,  when we do not want to give out sensitive parts of our code implementation and this is where data abstraction came.

```python
from abc import ABC,abstractmethod
class BankApp:
  def database(self):
    print('Connected to database')

  @abstractmethod
  def security(self):
    pass
```

```python
class BankApp(ABC):
  def database(self):
    print('Connected to Database')

  @abstractmethod
  def security(self):
    pass
```

```python
class mobileApp(BankApp):
  def mobile_login(self):
    print('Login into mobile')
```

```python
mob = mobileApp()
```

```python
class mobileApp(BankApp):
  def mobile_login(self):
    print('Login into mobile')
  def security(self):
    print('Mobile Security')
```

```python
mob=mobileApp()
```

```python
mob.mobile_login()
```

```python
mob.security()
```
